```python
## Exemplo de uso Autoviz
from autoviz.AutoViz_Class import AutoViz_Class  #Instantiate the AutoViz class
AV = AutoViz_Class()
%matplotlib inline
dft = AV.AutoViz("../datasets/bureau.csv") 
```

# Show grafico

```python
## 
AV = AutoViz_Class()
filename = ""
sep = ","
dft = AV.AutoViz(
    filename,
    sep=",",
    depVar="SK_ID_CURR",
    dfte=df_bureau,
    header=0,
    verbose=0,
    lowess=False,
    chart_format="svg",
    max_rows_analyzed=150000,
    max_cols_analyzed=300,
)
```
